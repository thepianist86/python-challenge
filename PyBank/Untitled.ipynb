```python
import os
import csv
```

```python
fpath = os.path.join("Resources", "budget_data.csv")

net_tot_amt = 0
months = 0
change = 0
aver_change = 0
first = 0
second = 0
```

```python
count = 0
change_tot = 0
with open(fpath) as bd_csv:
    bd_reader = csv.reader(bd_csv, delimiter=",")
    next(bd_reader)
    
    for r in bd_reader:
        if count == 0:
            first = int(r[1])
        else:         
            second = int(r[1])
            change = second - first
            print(change)
            change_tot += change
            first = second
        count += 1
    aver_change = change_tot/months
    
```
